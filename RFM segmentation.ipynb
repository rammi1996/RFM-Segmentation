```python
# RFM Analysis 
```

```python
df.head()
```

```python
# create a new column
df['total']=df['Price']*df['Quantity']
```

```python
df.InvoiceDate.max()
```

```python
 #Assuming 'InvoiceDate' is stored as a string in the format 'MM/DD/YYYY HH:MM'
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M', errors='coerce')

# Check for missing or invalid datetime values
invalid_dates = df[df['InvoiceDate'].isnull()]
if not invalid_dates.empty:
    print("Warning: There are missing or invalid datetime values.")

# Adding three days to the maximum invoice date
analysis_date = df['InvoiceDate'].max() + pd.DateOffset(days=3)

# Creating a new column 'recency'
df['recency'] = (analysis_date - df['InvoiceDate']).dt.days
```

```python
df.head()
```

```python
# Frequency
frequency=df.groupby('Customer ID')['Invoice'].count().reset_index()
frequency.rename(columns={'Invoice':'frequency'},inplace=True)
df=df.merge(frequency, on='Customer ID',how='left')
```

```python
# calculate monetory
monetary=df.groupby('Customer ID')['total'].sum().reset_index()
monetary.rename(columns={'total':'monetary'},inplace=True)
df=df.merge(monetary, on='Customer ID',how='left')
```

```python
df.head()
```

```python
# bins
recency_scores=[5,4,3,2,1]
frequency_scores=[1,2,3,4,5]
monetary_scores=[1,2,3,4,5]

df['recency_scores']=pd.cut(df['recency'],bins=5,labels=recency_scores)
df['frequency_scores']=pd.cut(df['frequency'],bins=5,labels=frequency_scores)
df['monetary_scores']=pd.cut(df['monetary'],bins=5,labels=monetary_scores)
```

```python
df.info()
```

```python
# convert RFM _scores into int data type 

df['recency_scores']=df['recency_scores'].astype(int)
df['frequency_scores']=df['frequency_scores'].astype(int)  
df['monetary_scores']=df['monetary_scores'].astype(int)   
   
```

```python
# RFM score ''
df['rfm_score']=df['recency_scores']+df['frequency_scores']+df['monetary_scores']
segment_label=['low_value','mid_value','high_value']
df['value_segment']=pd.qcut(df['rfm_score'],q=3,labels=segment_label)
```

```python
df.head()
```

```python
df.rfm_score.nunique()
```

```python
df.rename(columns={'value_segment':'rfm_segment'},inplace=True)
```

```python
df.head()
```

```python
df['customer_segment']=np.where(df['rfm_score']>=14 ,'champion customer',
                                np.where(df['rfm_score']>=12,'loyal customer',
                                         np.where(df['rfm_score']>=10,'potentional loyalist customer',
                                                 np.where(df['rfm_score']>=8,'promising customer',
                                                         np.where(df['rfm_score']>=6,'new customer',
                                                                 np.where(df['rfm_score']>=4,'at risk customer',
                                                                         'lost customer'))))))
```
